## K Folds Cross Validation

### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split , cross_val_score
from sklearn.metrics import accuracy_score

%matplotlib inline

pd.set_option('max_colwidth',999)
pd.set_option('display.max_columns',999)
pd.set_option('display.max_rows',999)
```

### Importing the data

```python
df = pd.read_csv('/home/parth/AppliedAI/datasets/knn/knn/demo_data/3.concertriccir2.csv',header=None)
df.head()
```

### Data Preprocessing

```python
x = df[[0,1]].values
y = df[2].values
```

```python
x_train , x_test , y_train , y_test  = train_test_split(x,y,test_size=0.30, random_state=42)
```

```python

k_list = np.arange(1,50,2)
cv_scores = []

for k in k_list:
    knn = KNeighborsClassifier(n_neighbors=k,n_jobs=-1)
    scores = cross_val_score(knn, x_train, y_train, cv = 3, scoring='accuracy')
    cv_scores.append(scores.mean())
```

```python
error = np.array(cv_scores) 
error = 1 - error
```

```python
plt.figure(figsize=(10,10))
plt.plot(k_list,error)
for x,y in zip(k_list,np.round(error,3)):
    plt.text(x,y,'({},{})'.format(x,y))
plt.plot(k_list,cv_scores)
for x,y in zip(k_list,np.round(cv_scores,3)):
    plt.text(x,y,'({},{})'.format(x,y))
plt.show()
```

```python
optimal_k = k_list[error.argmin()]
optimal_k
```

```python
knn_optimal = KNeighborsClassifier(n_neighbors=optimal_k,n_jobs=-1)
knn_optimal.fit(x_train,y_train)
pred = knn_optimal.predict(x_test)
acc = accuracy_score(y_test , pred )
```

```python
print(acc * 100)
```
