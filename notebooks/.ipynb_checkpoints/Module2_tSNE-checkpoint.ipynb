# Applying t-SNE on MNIST dataset

## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

## Loading the Dataset

```python
df = pd.read_csv('/home/parth/AppliedAI/datasets/mnist/train.csv')
df.head()
```

```python
df_label  = df['label']
df.drop('label',axis=1,inplace=True)
```

```python
print(df.head())
print(df_label.head())
```

## Column Standardization of the dataset

```python
df_standard = StandardScaler().fit_transform(df)
```

```python
df_standard
```

### Applying TSNE with default values and for all the data points

```python
tsne = TSNE(n_components=2,random_state=0)
tsne_data = tsne.fit_transform(df_standard)
tsne_data
```

### Combining the labels with the tsne_data

```python
tup= (tsne_data.T,df_label)
final_data = np.vstack(tup)
```

### Making the dataframe of the Matrix

```python
df_final = pd.DataFrame(final_data.T,columns=['v1','v2','label'])
df_final.head()
```

## Plotting the 1st TNSE 

```python
plt.figure()
sns.FacetGrid(df_final,hue='label',height=8).map(sns.scatterplot,'v1','v2').add_legend()
plt.show()
```

### function for solving the problem altogether

```python
def tsne_auto(data,iterations=1000,perplextiy = 30,data_standardized=None,random_state=0):
    n_components=2
    if data_standardized is None:
        data = StandardScaler().fit_transform(data) 
    tsne_data = TSNE(n_components= n_components,
                     random_state=  random_state,perplexity= perplextiy,
                     n_iter=iterations).fit_transform(data)
    df_final = pd.DataFrame(np.vstack((tsne_data.T,df_label)).T
                            ,columns=['v1','v2','label'])
    plt.figure()
    sns.FacetGrid(df_final,
                  hue='label',
                  height=8).map(
        sns.scatterplot,
        'v1','v2'
    ).add_legend()
    plt.show()
    return
```

```python
tsne_auto(df_standard,iterations=5000,perplextiy=50,data_standardized=True)
```
