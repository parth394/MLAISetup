## Random Sampling

### Imports 

```python
import random
from sklearn import datasets
# Printing a random number between 0 and 1
print(random.random())
```

### Importing the dataset 

```python

iris = datasets.load_iris()
d = iris.data
d.shape
```

### Random Sampling Code

```python
n = 150 
m = 30
p = m/n
random.seed(0)
sample_data = []
for i in range(0,n):
    # probabilty of random value generated is less than 0.2 is 20%
    temp = random.random()
    if temp <= p:
        sample_data.append(d[i,:])
print(len(sample_data))
```

The size of sample data will not always be 30 it will be roughly around 30

## Emperical Bootstrapping For Getting Confidence Interval of Sample

## Imports required

```python
import numpy as np
from sklearn.utils import resample
import matplotlib.pyplot as plt
```

### Creating Sample

```python
S = np.array([182,162,158,172,168,150,171,183,165,176])
```

k = No. of sample we need to create from the original sample

m = length of the new sample

```python
k = 1000
m = len(S)

sample_median = []
for iteration in range(0,k):
    s = resample(S,n_samples=m)
    median = np.median(s)
    sample_median.append(median)
alpha = 0.95
lower_percentile = ((1.0-alpha)/2.0) * 100
lower_bound = np.percentile(sample_median,lower_percentile)
upper_percentile = (alpha+((1-alpha)/2.0))*100
upper_bound = np.percentile(sample_median,upper_percentile)
```

```python
plt.figure(figsize=(8,8))
plt.hist(sample_median)
plt.xlabel('Median of the samples')
plt.title('''The Confidence Interval with 95% probability is
 [{} and {}]'''.format(lower_bound,upper_bound))
plt.show()
```

## K-S Test

### Imports

```python
import numpy as np
import seaborn as sns
from scipy import stats
from matplotlib import pyplot as plt
```

### Performing K-S Test

```python
x = stats.norm.rvs(size=1000)
sns.set_style('whitegrid')
sns.kdeplot(x,bw=0.5)
plt.show()
```

```python
stats.kstest(x,'norm')
```

```python
y = np.random.uniform(0,1,10000)
sns.kdeplot(y,bw=0.1)
plt.show()
```

```python
stats.kstest(y,'norm')
```
