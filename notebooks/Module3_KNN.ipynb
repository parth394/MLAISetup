### Imports 

```python
import pandas as pd 
import numpy as np
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

%matplotlib inline
sns.set(style="whitegrid", color_codes=True)

pd.set_option('max_colwidth',999)
pd.set_option('display.max_columns', 999)
pd.set_option("display.max_rows",999)
```

### Creating a function Knn and plotting

```python
def knn_comparison(data, n_neighbors = 15):
    x = data[[0,1]].values
    y = data[2].values
    
#     Grid Cell Size
    h = 0.02
#     Creating color for the grid
    cmap_light = ListedColormap(['#FFAAAA',  '#AAAAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#0000FF'])

#     Creating the Knn classification and fitting data to the model
    knn = KNeighborsClassifier(n_neighbors=n_neighbors , n_jobs=-1)
    knn.fit(x,y)
    
#     Creating min max points for the mesh
    x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
    y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
    
#     Creating the mesh grid for the data
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

#     Creating the testing value set for the classifier to predict
    test = np.c_[xx.ravel(),yy.ravel()]

#     Predicting
    prediction = knn.predict(test)
    
#     Reshaping the prediction
    prediction = prediction.reshape(xx.shape)
    
#     Plotting the mesh along with the data.
    plt.figure()
    plt.pcolormesh(xx, yy, prediction, cmap=cmap_light)
    plt.scatter(x[:, 0], x[:, 1], c=y, cmap=cmap_bold)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title('K value = '+str(n_neighbors))
    plt.show()
```

### Reading the datasets

```python
df = pd.read_csv('/home/parth/AppliedAI/datasets/knn/knn/demo_data/1.ushape.csv',header=None)
knn_comparison(df,1)
knn_comparison(df,5)
knn_comparison(df,15)
knn_comparison(df,30)
knn_comparison(df,50)
```
